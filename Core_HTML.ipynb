4. Create an ordered list of HTML tags. Each list item must include the tag name and some information about
the tag.

```python
<!DOCTYPE html>
<html>
<head>
    <title>HTML Tags List</title>
</head>
<body>
    <h1>Ordered List of HTML Tags</h1>
    <ol>
        <li>
            <strong>&lt;html&gt;</strong> - The root element of an HTML page. It wraps all the content on the web page.
        </li>
        <li>
            <strong>&lt;head&gt;</strong> - Contains meta-information about the document such as the title, links to CSS, and scripts.
        </li>
        <li>
            <strong>&lt;title&gt;</strong> - Sets the title of the webpage, which appears in the browser tab.
        </li>
        <li>
            <strong>&lt;body&gt;</strong> - Contains the visible content of the web page such as text, images, and links.
        </li>
        <li>
            <strong>&lt;h1&gt; to &lt;h6&gt;</strong> - Heading tags. <code>&lt;h1&gt;</code> is the largest, and <code>&lt;h6&gt;</code> is the smallest.
        </li>
        <li>
            <strong>&lt;p&gt;</strong> - Paragraph tag. Used to define a block of text.
        </li>
        <li>
            <strong>&lt;a&gt;</strong> - Anchor tag. Used to create hyperlinks using the <code>href</code> attribute.
        </li>
        <li>
            <strong>&lt;img&gt;</strong> - Image tag. Displays an image using the <code>src</code> and <code>alt</code> attributes.
        </li>
        <li>
            <strong>&lt;ul&gt;</strong> - Unordered list. Used with <code>&lt;li&gt;</code> tags to create bullet point lists.
        </li>
        <li>
            <strong>&lt;ol&gt;</strong> - Ordered list. Used with <code>&lt;li&gt;</code> tags to create numbered lists.
        </li>
        <li>
            <strong>&lt;div&gt;</strong> - Division tag. A generic container used for layout and grouping content.
        </li>
        <li>
            <strong>&lt;span&gt;</strong> - Inline container used to style parts of text.
        </li>
        <li>
            <strong>&lt;form&gt;</strong> - Used to create input forms. Often includes <code>&lt;input&gt;</code>, <code>&lt;textarea&gt;</code>, and <code>&lt;button&gt;</code>.
        </li>
        <li>
            <strong>&lt;input&gt;</strong> - Accepts user input. Type can be text, radio, checkbox, etc.
        </li>
        <li>
            <strong>&lt;button&gt;</strong> - Represents a clickable button, often used to submit forms.
        </li>
    </ol>
</body>
</html>
```
